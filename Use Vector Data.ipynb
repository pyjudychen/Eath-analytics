## What is Vector Data?
**Vector data** are composed of discrete geometric locations (x, y values) known as vertices that define the “shape” of the spatial object. 

![Vector Data](https://www.earthdatascience.org/images/earth-analytics/spatial-data/points-lines-polygons-vector-data-types.png)

## What Data are Stored in Spacial Vector Formats?
Some examples of data that often are provided in a vector format include:

* Census data including municipal boundaries
* Roads, powerlines and other infrastructure boundaries
* Political boundaries
* Building outlines
* Water bodies and river systems
* Ecological boundaries
* City locations
* Object locations including plots, stream gages, and building locations

##  Geoprocessing Vector Data Geoprocessing in Python: Clip Data
Sometimes you have spatial data for a larger area than you need to process. For example you may be working on a project for your state or country. But perhaps you have data for the entire globe.

You can clip the data spatially to another boundary to make it smaller. Once the data are clipped, your processing operations will be faster. It will also make creating maps of your study area easier and cleaner.

![Clip Data](https://www.earthdatascience.org/images/earth-analytics/spatial-data/vector-clip.png)

Below you do the following:

1. Subset the countries layer to just the boundary of the United State of America
2. Plot the data to look at the newly subsetted data!
3. Clip the cities data to only include cities that fall within the boundary of the United States

```python
# Subset the country data to just a single country
united_states_boundary = country.loc[country['SOVEREIGNT'] == 'United States of America']

# Notice in the plot below, that the only boundary for the USA is in the new variable
f, ax = plt.subplots(figsize = (10, 6))
united_states_boundary.plot(ax = ax)
plt.show()
```

```python
# Clip the city data to the USA boundary
# Note -- this oepration may take some time
city_in_usa = gpd.clip(city, united_states_boundary)

# Plot final clipped data
f, ax = plt.subplots()
united_states_boundary.plot(ax = ax, color = 'white', edgecolor = 'grey')
city_in_usa.plot(ax = ax, markersize = 2)
ax.set(title = 'Cities clipped to the USA Boundary')
plt.show()
```
