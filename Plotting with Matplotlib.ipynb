## Overviewing of plotting with Matplotlib

**Matplotlib** uses an object oriented approach to plotting. This means that plots can be built step-by-step by adding new elements to the plot.

There are two primary objects associated with a matplotlib plot:

* Figure object: the overall figure space that can contain one or more plots.
* Axis object: the individual plots that are rendered within the figure.

## Create plots using Matplotlib
You can change the size of your figure using the argument figsize to specify a width and height for your figure:
 
    figsize = (width, height)

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (10, 6))
```

## Multiplot figures

You will need to provide new arguments to plt.subplots for the layout of the figure: number of rows and columns:

    plt.subplots(1, 2)

```python
# Figure with 2 plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 6))
```

```python
# Figure with 2 plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10, 6))
```

## Customize plots using Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import data
monthly_precip_CO = pd.read_csv('./data/monthy_preciptation_CO.csv')
monthly_precip_CO.head()
```

```python
# Define plot size
fig, precip = plt.subplots(figsize = (10, 6))
month = monthly_precip_CO['Months'].values
monthly_precip_CO = monthly_precip_CO['Monthy_precip'].values

# Define x, y axes
precip.plot(month, monthly_precip_CO)
plt.show()
```

## Create different types of plots: scatter and bar plots
You may have noticed that by default, ax.plot creates the plot as a **line** plot (meaning that all of the values are connected by a continuous line across the plot).

You can also use the ax object to create:

* Scatter plots (using ax.scatter): values are displayed as individual points that are not connected with a continuous line.
* Bar plots (using ax.bar): values are displayed as bars with height indicating the value at a specific point.

```python
fig, precip = plt.subplots(figsize = (10, 6))

precip.scatter(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values)
plt.show()
```

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.bar(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values)
plt.show()
```

## Customize plot title and axes labels
You can customize and add more information to your plot by adding a plot title and labels for the axes using the title, xlabel, ylabel arguments within the *ax.set()* method:

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.plot(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values)
precip.set(title = 'Average Monthly Precipitation in Boulder, CO', 
      xlabel = 'Month', 
      ylabel = 'Preciptation (inches)')
plt.show()
```

### Multi-line titles and labels

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.plot(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values)
precip.set(title = 'Average Monthly Precipitation \n in Boulder, CO', 
      xlabel = 'Month', 
      ylabel = 'Preciptation \n (inches)')
plt.show()
```

### Rotate labels

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.plot(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values)
precip.set(title = 'Average Monthly Precipitation \n in Boulder, CO', 
      xlabel = 'Month', 
      ylabel = 'Preciptation \n (inches)')

plt.setp(precip.get_xticklabels(), rotation = 45)     
plt.show()
```

## Custom markers in line and scatter plots
You can change the point marker type in your line or scatter plot using the argument **marker =** and setting it equal to the symbol that you want to use to identify the points in the plot.

|Marker symbol|description|
|:------------|:----------|
|.|point|
|,|pixel|
|o|circle|
|v|triangle_down|
|^|triangle_up|
|<|triangle_left|
|>|triangle_right|

You can customize the color of your plot using the color argument and setting it equal to the color that you want to use for the plot. A list of some of the base color options available in matplotlib is below:
* b: blue
* g: green
* r: red
* c: cyan
* m: magenta
* y: yellow
* k: black
* w: white
    
For more colors, visit the [matplotlib documentation](https://matplotlib.org/stable/gallery/color/named_colors.html) on color.

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.scatter(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values, marker = ',')
precip.set(title = 'Average Monthly Precipitation \n in Boulder, CO', 
      xlabel = 'Month', 
      ylabel = 'Preciptation \n (inches)')

plt.show()
```

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.plot(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values, marker = 'o')
precip.set(title = 'Average Monthly Precipitation \n in Boulder, CO', 
      xlabel = 'Month', 
      ylabel = 'Preciptation \n (inches)')

plt.show()
```

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.plot(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values, marker = 'o', color = 'magenta')
precip.set(title = 'Average Monthly Precipitation \n in Boulder, CO', 
      xlabel = 'Month', 
      ylabel = 'Preciptation \n (inches)')

plt.show()
```

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.bar(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values, color = 'darkblue')
precip.set(title = 'Average Monthly Precipitation \n in Boulder, CO', 
      xlabel = 'Month', 
      ylabel = 'Preciptation \n (inches)')

plt.show()
```

### Set Color Transparency
You can also adjust the transparency of color using the **alpha =**  argument, with values closer to 0.0 indicating a higher transparency.

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.bar(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values, color = 'darkblue', alpha = 0.3)
precip.set(title = 'Average Monthly Precipitation \n in Boulder, CO', 
      xlabel = 'Month', 
      ylabel = 'Preciptation \n (inches)')

plt.show()
```

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.bar(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values, 
           color = 'cyan', edgecolor = 'darkblue')
precip.set(title = 'Average Monthly Precipitation \n in Boulder, CO', 
      xlabel = 'Month', 
      ylabel = 'Preciptation \n (inches)')

plt.show()
```

When using scatter plots, you can also assign each point a color based upon its data value using the **c** and **cmap** arguments.

The **c** argument allows you to specify the sequence of values that will be color-mapped, while **cmap** allows you to specify the color map to use for the sequence.
To see a list of color map options, visit the  [matplotlib documentation](https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html)on colormaps.

```python
fig, precip = plt.subplots(figsize = (10, 6))
precip.scatter(monthly_precip_CO['Months'].values, monthly_precip_CO['Monthy_precip'].values, 
               c = monthly_precip_CO['Monthy_precip'].values, cmap = 'YlGnBu')
precip.set(title = 'Average Monthly Precipitation \n in Boulder, CO', 
      xlabel = 'Month', 
      ylabel = 'Preciptation \n (inches)')

plt.show()
```

## Customize Matplotlib Dates Ticks on X-axis

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Date time conversion registration
from pandas.plotting import register_matplotlib_converters
```

```python
data_path = './data/colorado-flood/precipitation/805325-precip-dailysum-2003-2013.csv'
boulder_daily_precip = pd.read_csv(data_path, 
                                  parse_dates = ['DATE'], 
                                  na_values = ['999.99'], 
                                  index_col = ['DATE'])
boulder_daily_precip.head()
```

```python
# Subset data
precip_boulder_AugOct = boulder_daily_precip['2013-08-15':'2013-10-15']
precip_boulder_AugOct.head()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax.plot(precip_boulder_AugOct.index.values, 
       precip_boulder_AugOct['DAILY_PRECIP'].values, 
       '-o', color = 'purple')
ax.set(xlabel = 'Date', 
       ylabel = 'Precipitation (inches)', 
       title = 'Daily Precipitation \nBoulder, Colorado 2013')

ax.xaxis.set_major_locator(mdates.WeekdayLocator (interval = 2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax.bar(precip_boulder_AugOct.index.values, 
       precip_boulder_AugOct['DAILY_PRECIP'].values, 
       color = 'purple', alpha = 0.5)
ax.set(xlabel = 'Date', 
       ylabel = 'Precipitation (inches)', 
       title = 'Daily Precipitation \nBoulder, Colorado 2013')

# Define date fromat
date_form = mdates.DateFormatter('%m/%d')
ax.xaxis.set_major_formatter(date_form)

# Make the ticks fall once every other week (interval = 2)
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval = 2))
```
